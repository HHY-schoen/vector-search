```python
# TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer

# å‡è¨­æˆ‘å€‘æœ‰ä¸‰å€‹æ–‡æª”
documents = [
    "I love machine learning and deep learning.",
    "Machine learning is great for AI applications.",
    "Deep learning improves AI performance."
]

# å»ºç«‹ TF-IDF å‘é‡å™¨
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents)

# è½‰æ›ç‚ºå¯è®€çš„çŸ©é™£æ ¼å¼
print(tfidf_matrix.toarray())

# é¡¯ç¤ºå°æ‡‰çš„è©å½™
print(vectorizer.get_feature_names_out())

# é¡¯ç¤ºäº†æ¯å€‹å–®è©åœ¨ä¸åŒæ–‡æª”ä¸­çš„ TF-IDF å€¼ï¼Œå€¼è¶Šå¤§ä»£è¡¨è©²è©å°è©²æ–‡æª”çš„é‡è¦æ€§è¶Šé«˜ã€‚
```

```python
# TF-IDF + å‘é‡æœå°‹
from sklearn.feature_extraction.text import TfidfVectorizer
from sentence_transformers import SentenceTransformer, util
import numpy as np

# æ–‡å­—æ•¸æ“šåº«
documents = [
    "æ©Ÿå™¨å­¸ç¿’æ˜¯äººå·¥æ™ºæ…§çš„ä¸€éƒ¨åˆ†ã€‚",
    "æ·±åº¦å­¸ç¿’æ˜¯ä¸€ç¨®æ©Ÿå™¨å­¸ç¿’æ–¹æ³•ã€‚",
    "è‡ªç„¶èªè¨€è™•ç†è®“æ©Ÿå™¨ç†è§£èªè¨€ã€‚",
    "æœå°‹å¼•æ“ä¾è³´æ–¼é—œéµå­—åŒ¹é…å’Œèªç¾©æœå°‹ã€‚"
]

# ğŸ”¹ TF-IDF é—œéµå­—æœå°‹
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(documents)

# ğŸ”¹ å‘é‡æœå°‹ï¼ˆBERTï¼‰
model = SentenceTransformer('all-MiniLM-L6-v2')  # è¼•é‡ç´šçš„èªæ„å‘é‡ï¼ˆSentence Embeddingï¼‰æ¨¡å‹
vector_embeddings = model.encode(documents)

# ğŸ”¹ ç”¨æˆ¶æŸ¥è©¢
query = "AIå­¸ç¿’"
query_embedding = model.encode(query)

# è¨ˆç®—å‘é‡ç›¸ä¼¼åº¦
cosine_scores = util.pytorch_cos_sim(query_embedding, vector_embeddings)

# è¼¸å‡ºæœ€ç›¸ä¼¼çš„çµæœ
top_result = np.argmax(cosine_scores.numpy())
print("æœ€ä½³åŒ¹é…çµæœ:", documents[top_result])
```
